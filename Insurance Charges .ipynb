Converting the categorical values

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
sex_encoded_data = encoder.fit_transform(df['sex'])
df['sex'] = sex_encoded_data

smoker_encoded_data = encoder.fit_transform(df['smoker'])
df['smoker'] = smoker_encoded_data

region_encoded_data = encoder.fit_transform(df['region'])
df['region'] = region_encoded_data
```

## Building a model

### Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('charges', axis = 1)
```

```python
X
```

```python
Y = df['charges']
```

```python
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=0, test_size=0.2)
```

```python
x_train
```

## Importing the linear model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Now, we train the model

```python
model.fit(x_train, y_train)
```

The model is now ready for predictions

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

### Checking the accuracy of the model

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

## Assumptions

#### 1. Linearity

Independent variables should show a linear relationship with dependent varibale

#### 2. No endogeneity

#### 3. Normality and Homoscedasticity

For homoscedasticity, we're checking the deviation of data points from the line

```python
residuals = y_test-y_pred
```

```python
residuals
```

we check the mean of the residuals

```python
mean_res = np.mean(residuals)
mean_res
```

```python
plot = sns.scatterplot(x= y_pred, y = residuals)
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.ylim(-20000, 20000)

plot= sns.lineplot(x = [0, 40000], y = [0, 0])
```

We find no heteroscedasticity because variances are equal across all data points.

To prove it we use any hypothesis testing, in this case Goldfeld Quandt Test






null hypo - variances between mean residuals and residuals follows homoscedasticity

null hypo - variances between mean residuals and residuals follows heteroscedasticity

```python
from statsmodels.stats.diagnostic import het_goldfeldquandt

gq_test = het_goldfeldquandt(residuals, x_test)

gq_test
```

```python
sns.distplot(residuals)
```

#### 4. No autocorrelation

```python
plot = sns.lineplot(x = y_pred, y = residuals, marker = 'o', color = 'green')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.ylim(-20000, 20000)

plot= sns.lineplot(x = [0, 40000], y = [0, 0])
```

```python
#Autocorrelation
import statsmodels.api as sm

sm.graphics.tsa.plot_acf(residuals)
```
